Here in the dataset 'mileage(kmpl)', 'engine(cc)', 'max_power(bhp)', 'torque(Nm)' has missing values so we need to impute those missing values with median/mean/mode. Here all the values are numerical so I use median to impute the missing values 

```python
cars['mileage(kmpl)']=cars['mileage(kmpl)'].fillna(cars['mileage(kmpl)'].median())

cars['max_power(bhp)']=cars['max_power(bhp)'].fillna(cars['max_power(bhp)'].median())

cars['engine(cc)']=cars['engine(cc)'].fillna(cars['engine(cc)'].median())

cars['torque(Nm)']=cars['torque(Nm)'].fillna(cars['torque(Nm)'].median())
```

```python
from sklearn.preprocessing import StandardScaler   
scaler = StandardScaler()
sc_x = scaler.fit_transform(x)
pd.DataFrame(sc_x)
```

```python
## variance_inflation_factor of Independent variables
var = sc_x

from statsmodels.stats.outliers_influence import variance_inflation_factor
import pandas as pd

# Assuming sc_x and x are already defined
var = sc_x
vif = pd.DataFrame()
vif["variance_inflation_factor"] = [variance_inflation_factor(var, i) for i in range(var.shape[1])]
vif["features"] = x.columns

# Display the VIF dataframe
vif
```

A variance inflation factor (VIF) is a measure of the amount of multicollinearity in regression analysis. Multicollinearity exists when there is a correlation between multiple independent variables in a multiple regression model. Here all the variables shows low VIF score that represents there is no multicollinearity among the independent variables
